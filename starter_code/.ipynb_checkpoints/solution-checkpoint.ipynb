### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
cursor = db.companies.find({"name": "Babelgum"})
print(list(map(lambda r:r["name"],list(cursor))))
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.


```python
cursor2 = db.companies.find({"number_of_employees": {'$gt':5000}}).sort("number_of_employees",pymongo.ASCENDING).limit(20)
list(cursor2)
```

3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
cursor3 = db.companies.find({ '$and': [{'founded_year': { '$gte': 2000} }, {'founded_year': {'$lte': 2005 } } ]})
print(list(map(lambda r :(r["name"], r['founded_year']), list(cursor3))))
```
