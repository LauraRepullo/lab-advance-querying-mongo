```python
import pymongo
MongoClient = pymongo.MongoClient
import time
client = MongoClient()
import pandas as pd
import re
db = client.companies
```

# Answers

### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
cursor = db.companies.find({"name": "Babelgum"})
print(list(map(lambda r:r["name"],list(cursor))))
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.


```python
cursor2 = db.companies.find({"number_of_employees": {'$gt':5000}}).sort("number_of_employees",pymongo.ASCENDING).limit(20)
list(cursor2)
```

3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
cursor3 = db.companies.find({ '$and': [{'founded_year': { '$gte': 2000} }, {'founded_year': {'$lte': 2005 } } ]})
print(list(map(lambda r :(r["name"], r['founded_year']), list(cursor3))))
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields

```python
cursor4 = db.companies.find({ '$and': [{"ipo.valuation_amount":{"$gte": 100000000}}, {'founded_year': {'$lte': 2010 } } ]})
print(list(map(lambda r :(r["name"], r['ipo']), list(cursor4))))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
cursor5 = db.companies.find({'$and':[{"number_of_employees": {'$gte':1000}}, {'founded_year': {'$lte': 2005}}]}).sort("number_of_employees",pymongo.DESCENDING).limit(10)
list(cursor5)
```
